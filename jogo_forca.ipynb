## Classe Principal

```python
# -*- coding: utf-8 -*-
class JogoForca:
    def __init__(self, word):
        self.word = word
        self.cur_letter = ""
        self.status_word = "_____"
        self.corrects = []
        self.errors = 0
        self.executions = 0

    def contains_latter(self, letter):
        self.cur_letter = letter
        if self.word.__contains__(letter):
            if not self.corrects.__contains__(letter):
                self.corrects.append(letter)
            return True
        return False

    def show_word(self):
        status = ''
        for l in self.word:
            if l.__eq__(self.cur_letter):
                status += self.cur_letter
            elif self.corrects.__contains__(l):
                status += l
            else:
                status += '_'
        self.status_word = status
        print('Palavra:', self.status_word)

    def isFinished(self):
        if self.status_word.__eq__(self.word):
            return True
        else:
            return False
```
