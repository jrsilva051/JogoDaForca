## Posições do Tabuleiro

```python
tabuleiro = [
    '''
            +----+
                 |
                 |
                 |
                 |
        =============         
    ''',
    '''
            +----+
            |    |
            O    |
                 |
                 |
                 |
        =============         
    ''',
    '''
            +----+
            |    |
            O    |
           /|\   |
                 |
                 |
        =============         
    ''',
    '''
            +----+
            |    |
            O    |
           /|\   |
           / \   |
                 |
        =============         
    '''
]
```

## Classes e Funções

```python
# -*- coding: utf-8 -*-
import random
import os

class JogoForca:
    def __init__(self, word):
        self.word = word
        self.status_word = "_____"
        self.corrects = []
        self.errors = 0
        self.executions = 0

    def contains_latter(self, letter):
        self.executions += 1
        if self.word.__contains__(letter):
            print('Você acertou a letra:', letter)
            self.corrects.append(letter)
        else:
            print('Palavra não possúi a letra:', letter)
            self.errors += 1

    def show_word(self):
        status = ''
        for l in self.word:
            if self.corrects.__contains__(l):
                status += l
            else:
                status += '_'
        self.status_word = status
        print('Palavra:', self.status_word)

    def isFinished(self):
        if self.status_word.__eq__(self.word):
            return True
        else:
            return False

def random_word():
    with open('base_jogo.txt', 'rt') as f:
        data = []
        data_raw = f.readlines()
        for d in data_raw:
            data.append(d.replace('\n', ''))
        return random.choice(data)
    
```

## Principal

```python
def main():
    secret_word = random_word()
    game = JogoForca(secret_word)
    print('**** Jogo da Forca ***')
    
    while game.isFinished() == False and game.errors < 6:
        option_user = input('Digite uma letra:')
        game.contains_latter(option_user)
        print(tabuleiro[game.errors])
        game.show_word()
        print('\n')

    if game.isFinished() == True:
        print('Parabéns, você concluiu o desafio!')
    else:
        print('Não foi dessa vez, mas não desista!')

if __name__ == "__main__":
    main()
```
